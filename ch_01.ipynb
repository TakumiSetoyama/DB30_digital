# データベーススペシャリスト

## 1.データベースとは

### 1.1 データベースの基本
- データを1ヶ所で集中管理
 - 管理しやすい＞業務の効率化
- データのシステムから独立
 - システムごとにデータを管理する二度手間の削除
 - システムが必要に応じてデータベースにアクセスする

### 1.1.2 データモデル
**データモデルの種類**<br>
1. 概念データモデル
 - どのような内容のデータを格納するか
2. 論理データモデル
 - どのようにデータベースに実装するか
3. 物理データモデル
 - 物理配置などによる最適化

**主な論理データモデル**
- 階層型(ハイアラキカル)モデル
 - 親子関係
 - 子は複数の親を持てない
- ネットワークモデル
 - 子が複数の親を持てるように改良
- 関係(リレーショナル)モデル
 - テーブルとテーブル間の関係でデータを表す
 - 最も一般的
- オブジェクト指向モデル
 - データの型(クラス)を作成し、そのクラスからデータの数だけインスタンスを建てる
 - それぞれのクラスのインスタンス間で関連付けを行いデータを表す
 - オブジェクトと関係を対応付ける**O-Rマッピング**でRDBに格納できる
 
 
 **c.f.オブジェクト指向**<br>
データと操作を一体化して扱う考え方<br>
データ型の定義であるクラスをもとにオブジェクトの実体であるインスタンスを建てる<br> 
インスタンスとは実際のデータそのもの

### 1.1.3 関係モデル
リレーショナルとタプルで表現する<br>
属性$X,Y$の定義域(ドメイン)が$X\{x_1, x_2\}, Y\{y_1, y_2\}$で表されるとき<br>
その直積$X\cdot Y$<br>
(x_1, y_1)<br>
(x_1, y_2)<br>
(x_2, y_1)<br>
(x_2, y_2)<br>
のひとつひとつが行,組（タプル）である<br>
これらの要素のうち実在する組合せのみを取り出しもの(直積の部分集合)が関係(リレーショナル)であり表(テーブル)として表される<br>

| $X$ | $Y$ | 
| :---: | :---: |
| $x_1$ | $y_1$ |
| $x_2$ | $y_2$ |


### 1.1.4 データベースの3値理論
TrueとFalseとNULL

**NULLの意味**
1. unknown(不明・未知)
2. Inapplicable(非存在・適応不可)

## 1.2 システム開発とデータベース
データベースを活かすシステム開発

### 1.2.1 システム開発のアプローチ
1. システム中心アプロー(POA:Process Oriented Approach)
 - DFD(Data Flow Diagram)や状態遷移図でデータの流れ(データフロー)が表現される
 - DFDではプロセス(ソフトウェアの機能)間で渡されるデータを表す<br>
<!-- ![DFDデータフロー図の記載例](DFDデータフロー図の記載例.png "DFDデータフロー図の記載例") -->
2. データ中心アプローチ(DOA:Data Oriented Approach)←主にこれ
 - 業務で扱うデータに着目したアプローチ
 - 業務で扱うデータ全体をモデル化, データベースを設計
 - このデータベースを中心にシステムを設計ー＞システム間のデータの整合性・一貫性が担保
 - E-R図(Entity-Relationship Diagram)で実体(データ)と実体間の関係を表現する
3. オブジェクト指向アプローチ(OOA:Object Oriented Approach)
 - プログラムやデータをオブジェクトとみなす
 - クラス図やシーケンス図などのUML(Unified Modeling Language)で表す

### 1.2.2 データ中心アプローチ
データ中心アプローチのシステム開発の流れ

1. 開発する対象を絞り現実世界から**対象世界**を決定する
2. 対象世界をに対して**概念設計**(モデリングorデータモデリング)を行う
3. 概念設計(トップダウンア・プローチ, ボトムアップ・アプローチ)のにより<font color="Red">概念データモデル</font>が得られる
4. これに対してDBとユーザーやDB以外のシステムを結び付ける設計である**論理設計**を行い<font color="Red">論理データモデル</font>を作成
5. DBとDBMS(データベース管理システム)やハードウェアを結び付ける設計である**物理設計**を行い<font color="Red">物理データモデル</font>を作成

データモデルが3層に分かれていることでそれぞれのデータ独立が達成される<br>
例えば...
- DB以外のシステムでの変更があれば論理データモデルのみ変更すれば良い
- DBMSに変更があれば物理データモデルのみ変更すれば良い
- 概念データモデルはDBMSやシステムに依存しない机上のデータモデル
以上からわかるように変更による影響箇所の少ない変更に強いシステム構築

上の3層モデル以外にもデータベースを3層にわかる方法はある<br>
**３層スキーマアーキテクチャ**
1. 外部スキーマ
 - DBに関係ないプログラムやユーザーが使用するデータベースの記述
 - 「ビュー」とか
2. 概念スキーマ
 - エンティティやテーブル, テーブル間の関連の記述
3. 内部スキーマ
 - DBMSで使用するデータベースを物理的にどのように配置するかなどの具体的な実装のための記述
 - 「インデックス」とか

## 1.3 データ分析とデータベース

**OLTP**
- OnLine Transaction Processing
- 通常のシステムで行うトランザクションを中心とした処理方法


**OLAP**
- OnLine Analytical Processing
- 複雑で分析的な問い合わせに素早く答える処理方法
- 従来のOLTP(RDBなど)のデータのスナップショットを取り別のDBに移動する
- そのとき多次元データとして再構築して分析できるようにする
- この多次元データの集まりが**データウェアハウス**

**データウェアハウス**<br>
$def$「意思決定のための、目的別に編成され結合された<font color="Red">時系列で、削除や更新を行わない</font>データの集合体」


**データマイニング**
- Data Mining
- 大量のデータに統計学・パターン認識・機械学習などの解析を適応することで新たな知識を取り出すこと

### 1.3.2 AIで活用するためのデータベース

**ビックデータ**<br>
RDBMSだけでは扱いきれないデータ<br>
以下の３つの特徴(3V)を持つ
- Volume
 - データの量
 - 1台のサーバーでは処理しきれない
- Variety
 - データの種類
 - 種類多く頻繁に変更される場合テーブルの構造が固定されているRDBMSでは管理コストがかかる
- Veloocity
 - データの発生・処理の頻度・速度
 - トランザクション管理で整合性を保っているが速度が遅い
 - 処理の頻度が多いくクエリが連続して発行される場合コストがかかる
